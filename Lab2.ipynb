```python
import pandas as pd

data = {
    "Age": [25, 36, 30, 27, 38, 42, 34],
    "Salary": [42000, 50000, 45000, 43000, 51000, 62000, 48000]
}
df = pd.DataFrame(data)
```

```python
def min_max_normalize(column):
    return (column - column.min()) / (column.max() - column.min())
    
df['Normalized_Age'] = min_max_normalize(df['Age'])
df['Normalized_Salary'] = min_max_normalize(df['Salary'])
```

```python

print("Original DataFrame:")
print(df[['Age', 'Salary']])
print("\nNormalized DataFrame:")
print(df[['Normalized_Age', 'Normalized_Salary']])
```

```python
import pandas as pd

data = {
    "Age": [25, 36, 30, 27, 38, 42, 34],
    "Salary": [42000, 50000, 45000, 43000, 51000, 62000, 48000]
}
df = pd.DataFrame(data)
```

```python
def standardize(column):
    return (column - column.mean()) / column.std()

df['Standardized_Age'] = standardize(df['Age'])
df['Standardized_Salary'] = standardize(df['Salary'])
```

```python
print("Original DataFrame:")
print(df[['Age', 'Salary']])
print("\nStandardized DataFrame:")
print(df[['Standardized_Age', 'Standardized_Salary']])
```

```python
import pandas as pd
import numpy as np

data = {
    'First Score': [100, 90, np.nan, 95],
    'Second Score': [30, 45, 56, np.nan],
    'Third Score': [np.nan, 40, 80, 98]
}
```

```python
df = pd.DataFrame(data)
df_interpolated = df.interpolate(method='bfill')
```

```python
print("Original DataFrame:")
print(df)
print("\nDataFrame after Backward Interpolation:")
print(df_interpolated)
```
